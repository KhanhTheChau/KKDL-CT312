```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/Dry_Bean_Dataset2025.csv', encoding='utf-8')
df.head(5)
```

```python
df = df.drop(['ShapeFactor5.1'], axis=1)
df.head(5)
```

```python
# Kiểm tra giá trị null trong tập dữ liệu
df.isnull().sum()
```

```python
for column in df.columns:
    print(df[column].unique().tolist())
```

```python
# Loại bỏ các hàng chứa giá trị null trong tập dữ liệu
clear_null_df = df.dropna()
clear_null_df.isnull().sum()

y = clear_null_df['Class']
X = clear_null_df.drop(['Class', 'Name'], axis=1)

X.to_csv("./data/X_train.csv", index=False)
```

```python
columns_fixed = []
for col in X.columns:
    if not np.issubdtype(X[col].dtype, np.number):
        try:
            X[col] = X[col].astype(str).str.replace(',', '.', regex=False).astype(float)
            columns_fixed.append(col)
        except Exception as e:
            print(f"Không thể chuyển đổi cột '{col}'. Vẫn còn lỗi không phải dấu thập phân.")
            
print("Các cột đã được chuyển đổi:", columns_fixed)
```

```python
from sklearn.preprocessing import MinMaxScaler

for column in X.columns:
    scaler = MinMaxScaler()
    X[[column]] = scaler.fit_transform(X[[column]])
X.head(5)
```

```python
normalized_df = pd.concat([X, y], axis=1)
normalized_df.to_csv('./data/normalized_dry_bean_dataset.csv', index=False)
```

```python
df = pd.read_csv('./data/normalized_dry_bean_dataset.csv', encoding='utf-8')
df.head(5)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

X = df.drop(columns=['Class'])

pca = PCA(n_components=X.shape[1])
pca.fit(X)

loadings = pd.Series(pca.components_[0], index=X.columns)

abs_loadings = loadings.abs().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
abs_loadings.plot(kind='bar', color='darkblue')
plt.title('Mức độ Quan trọng của Cột đối với Thành phần Chính 1 (PC1)')
plt.ylabel('Trọng số Tuyệt đối (Absolute Loadings)')
plt.xlabel('Cột Tính năng')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
# plt.savefig('pca_loadings_pc1.png')
plt.show()
plt.close()
```

```python
explained_variance_ratio = pca.explained_variance_ratio_

# print("Tỷ lệ phương sai giải thích được (theo từng PC):\n", explained_variance_ratio)

plt.figure(figsize=(10, 6))
plt.plot(range(1, X.shape[1] + 1), explained_variance_ratio.cumsum(), marker='o', linestyle='--')
plt.title('Xác định Số lượng Thành phần Chính (k) Tối ưu bằng Phương sai Tích lũy')
plt.xlabel('Số lượng Thành phần Chính')
plt.ylabel('Phương sai Giải thích được Tích lũy')
plt.grid(True)
plt.show()
```
